```python
import numpy as np

hours = df['transaction_timestamp'].dt.hour.values  
#    → this is a NumPy array of length 11 000 000, very fast

# 2) build a 24-element lookup table
hour_labels = np.array(['Night'] * 24, dtype=object)
hour_labels[ 5:12] = 'Morning'    # 05:00–11:59
hour_labels[12:17] = 'Afternoon'  # 12:00–16:59
hour_labels[17:21] = 'Evening'    # 17:00–20:59
# hours 21–23, 0–4 remain 'Night'

# 3) index into it
df['time_of_day'] = hour_labels[hours]

df.drop(columns=['transaction_timestamp'], inplace=True)
```

```python
cities = df['merchant_city']

# 2. Get the unique city names as a NumPy array
cities.dropna(inplace=True)  # Remove any NaN values
unique_cities = cities.unique()

# 3. (Optional) Sort them and turn into a list for easy reading
unique_cities_sorted = sorted(unique_cities.tolist())

print(unique_cities_sorted) 

print(df['merchant_city'].value_counts())
```

```python
# 1. Define your ordered list of categories
categories = [
    "Clothing & Apparel",
    "Dining & Restaurants",
    "Electronics & Software",
    "Fuel & Service Stations",
    "General Retail & Department",
    "Grocery & Food Markets",
    "Hobby, Books, Sporting Goods",
    "Home Furnishings & Supplies",
    "Pharmacies & Health",
    "Services (Other)",
    "Travel & Transportation",
    "Unknown",
    "Utilities & Bill Payments"
]

# 2. Build the mapping dict
cat_to_code = {cat: idx+1 for idx, cat in enumerate(categories)}

# 3. Replace the column in-place
df['mcc_category'] = df['mcc_category'].map(cat_to_code)

# 4. Verify
print(df['mcc_category'].unique())
```

```python
# 1. Define the ordered list of your transaction types
types = [
    'POS',
    'BILL_PAYMENT',
    'ECOM',
    'ATM_WITHDRAWAL',
    'P2P_OUT',
    'P2P_IN',
    'SALARY'
]

# 2. Build a mapping dict (1-based codes)
type_to_code = {t: i+1 for i, t in enumerate(types)}

# 3. Replace the column in-place
df['transaction_type'] = df['transaction_type'].map(type_to_code)

# 4. Verify you now have codes 1–7 only
print(df['transaction_type'].unique())
print(df['transaction_type'].value_counts())
```

```python
import pandas as pd

# 1. Загрузить агрегированные фичи (customer-level)
from customer_segmentation_per_user import aggregate_features

# Укажите свой путь к исходному CSV
df = pd.read_csv('/Users/lithrur/Desktop/Coding projects/hackathon_2025_ai/dataset/dataset_cleaned.csv')
cust_feats = aggregate_features(df)

# 2. Собрать словарь с описаниями
rows = []
for col in cust_feats.columns:
    dtype = str(cust_feats[col].dtype)
    if col.endswith('_sum') or col.endswith('_mean') or col.endswith('_std'):
        base, stat = col.rsplit('_', 1)
        desc = {
            'sum':  'Сумма транзакций/значений',
            'mean': 'Среднее транзакций/значений',
            'std' : 'Стандартное отклонение транзакций/значений'
        }[stat] + f" для поля `{base}` на одного клиента"
        source = "groupby(card_id).agg"
    else:
        # колонки one-hot
        prefix, val = col.split('_', 1)
        desc = f"Число транзакций, где `{prefix}` = «{val}»"
        source = "one-hot encode + groupby(card_id).sum()"
    rows.append({
        "Field": col,
        "Type": dtype,
        "Description": desc,
        "Formula/Source": source
    })

data_dict = pd.DataFrame(rows)


data_dict.to_excel('data_dictionary.xlsx', index=False)

print("Data Dictionary создан:\n", data_dict.head(10))
```
